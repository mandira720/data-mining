```python
x=data.drop('Claimed',axis=1)
y=data.pop('Claimed')
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,train_label,test_label= train_test_split(x,y,test_size=.30,random_state=1,stratify=y)
```

```python
X_train.shape
```

```python
dt_model=DecisionTreeClassifier(criterion='gini')
```

```python
dt_model.fit(X_train,train_label)
```

```python
label=['NO','YES']
```

```python
plt.figure(figsize=(30,30))
tree.plot_tree(dt_model,max_depth=10,feature_names=list(X_train),class_names=label,fontsize=10);
plt.savefig("tree2.jpg")
```

```python
dt_mode= DecisionTreeClassifier(criterion='gini',max_depth=4,min_samples_leaf=20,min_samples_split=60)
```

```python
param_grid = {
    'criterion': ['gini'],
    'max_depth': [3,4,5,6],
    'min_samples_leaf': [10,20,30], 
    'min_samples_split': [30,60,90], 
}

dt_mode = DecisionTreeClassifier()

grid_search = GridSearchCV(estimator = dt_mode, param_grid = param_grid, cv = 13)
```

```python
grid_search.fit(X_train, train_label)
print(grid_search.best_params_)
best_grid = grid_search.best_estimator_
best_grid
```

```python
labels=['YES','NO']
```

```python
plt.figure(figsize=(20,20))
tree.plot_tree(best_grid,feature_names=list(X_train),class_names=labels,fontsize=15);
plt.savefig("tree.jpg")
```
